**Prediction on 'census income' using Logistic Regression, DecisionTree, RandomForest and Boosting models:**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("https://drive.google.com/uc?export=download&id=1sacUhRBfp8rO3Igm-I-6lpXYqYepnPBP")  #census-income.csv
df.head()
```

```python
#df.isnull().sum()
df.info()
```

```python
df.drop_duplicates(inplace=True) #removed 21
```

```python
df.info()
```

```python
#use one-hot encoding to avoid ML to assume state-gov workclass (1) is greater than private workclass (0) - if used label-encoder.
#dfe = pd.get_dummies(df,[col for col in df.columns if df[col].dtype=="object"],drop_first=True)  #drop_first to avoid multicolinearity (ask AI)
#Tree models need label-encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dfe = df.copy()
for col in dfe.columns:
  if dfe[col].dtype=="object":
    dfe[col] = le.fit_transform(dfe[col])

dfe.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import *

x=dfe.iloc[:,:-1]
y=dfe["annual_income"]
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=42)
lr = LogisticRegression()
lr.fit(xtrain,ytrain)
ypred = lr.predict(xtest)
accuracy_score(ytest, ypred)  #78%
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(xtrain,ytrain)
ypred = dt.predict(xtest)
accuracy_score(ytest, ypred) #81%
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(xtrain,ytrain)
ypred = rf.predict(xtest)
accuracy_score(ytest, ypred) #85%
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
xtrain = ss.fit_transform(xtrain)
xtest = ss.transform(xtest)
```

```python
from sklearn.ensemble import AdaBoostClassifier

ab = AdaBoostClassifier()
ab.fit(xtrain,ytrain)
ypred = ab.predict(xtest)
accuracy_score(ytest, ypred) #85%
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier()
gb.fit(xtrain,ytrain)
ypred= gb.predict(xtest)
accuracy_score(ytest, ypred)
```

```python
from xgboost import XGBClassifier

xgb = XGBClassifier()
xgb.fit(xtrain,ytrain)
ypred = xgb.predict(xtest)
accuracy_score(ytest, ypred)
```
